```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(8)
    M=len(points)
    for i in range(M):
        x=points[i,:]
        y=y1[i]
        for j in range(len(x)):
            s=[]
            for k in range(len(x)):
                s.append(x[k]*m[k])
            h=1/(1+ math.exp(-1*sum(s)))
            m_slope[j]+=(-1/M)*(y-h)*x[j]

    new_m=m-learning_rate*m_slope

    return new_m
```

```python
def gd(points,learning_rate,num_iterations):
    m=np.zeros(8)
    for i in range(num_iterations):
        m=step_gradient(points,learning_rate,m)
        print(i,"Cost:",cost(points,m))
    return m
```

```python
def cost(points,m):
    total_cost=0
    M=len(points)
    for i in range(M):
        x=points[i,:]
        y=y1[i]
        b=[]
        for j in range(len(m)):
            b.append (m[j]*x[j])
        h =1/(1+math.exp(-1*sum(b)))
        total_cost+=(-1/M)*(y*(math.log(h))+((1-y)*(math.log(1-h))))
    return total_cost
```

```python
def predicted(x3,m):
    y_p=[]
    for i in range(len(x3)):
        b=[]
        x=x3[i,:]
        for j in range(len(m)):
            b.append(m[j]*x[j])
        h=1/(1+math.exp(-1*sum(b)))
        if h>0.5:
            y_p.append(1)
        elif h<=0.5:
            y_p.append(0)
    return(y_p)
```

```python
def run():
    learning_rate=0.1
    num_iterations=500
    m =gd(x2, learning_rate,num_iterations)
    slope=m
    y_predicted=predicted(x3,slope)
    print(m)
    print(y_predicted)
    print(len(y_predicted))
    np.savetxt("y_titanic.csv",y_predicted,fmt='%.5f')
```
